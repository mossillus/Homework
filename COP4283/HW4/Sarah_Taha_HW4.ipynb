## Task 2: Altair
Create two different Altair charts using the same dataset.

```python
import altair as alt
import pandas as pd
import numpy as np

df = pd.read_csv('Cleaned Blue Dragon Drops.csv')
df.head()
```

```python
DropsPerTable = (
    alt.Chart(df)
    .mark_bar()
    .encode(
        x=alt.X("Table Title:N", title="Drop Table"),
        y=alt.Y("count():Q", title="Number of Drops"),
        color=alt.Color("Price Tiered:N", title="Price Tier"),
    )
    .properties(
        title="Number of Drops by Table and Price Tier",
    )
)

DropsPerTable
```

```python
DropsbyTiers = (
    alt.Chart(df)
    .mark_bar()
    .encode(
        x=alt.X("Price Tiered:N", title="Price Tier"),
        y=alt.Y("count():Q", title="Number of Drops"),
        color=alt.Color("Rarity Tier:N", title="Rarity Tier"),
    )
    .properties(
        title="Number of Drops by Price Tier and Rarity Tier",
        width=400,
        height=300
    )
)

DropsbyTiers
```
